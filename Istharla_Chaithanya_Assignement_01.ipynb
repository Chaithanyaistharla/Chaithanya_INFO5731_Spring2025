(b) (5 pts) For all the years, calculate the percentage of housing units which are vacant an occupied. Print the results in the following format:

Year Vacant Housing Units Occupied Housing Units

    2008  30%   70%                          
	2009 ----- -----

                              Year	Value
	Vacant Housing Units	  -----   -----

(d) (5 pts) Calculate the harmonic mean of the total housing units and print it out. You can find the information about harmonic mean here: https://ncalculators.com/statistics/harmonic-mean-calculator.htm

```python
import pandas as pd
from scipy.stats import hmean

fp = "/Assignment1_denton_housing.csv"
df = pd.read_csv(fp)
df_pivot = df.pivot(index="year", columns="title_field", values="value")
df_pivot = df_pivot.sort_index(ascending=False)
occ_diff = df_pivot["Occupied Housing Units"].diff().dropna()
occ_diff_sorted = occ_diff.sort_values(ascending=True)
print("Year-to-Year Difference in Occupied Housing Units:")
for year, diff in occ_diff_sorted.items():
    print(f"{year}: {diff}")

df_pivot["Vacant Percentage"] = (df_pivot["Vacant Housing Units"] / df_pivot["Total Housing Units"] * 100).round(2)
df_pivot["Occupied Percentage"] = (df_pivot["Occupied Housing Units"] / df_pivot["Total Housing Units"] * 100).round(2)
print("\nYear  Vacant Percentage  Occupied Percentage")
for year, row in df_pivot.iterrows():
    print(f"{year}  {row['Vacant Percentage']}%  {row['Occupied Percentage']}%")

max_year = df_pivot["Vacant Housing Units"].idxmax()
max_value = df_pivot["Vacant Housing Units"].max()
max_occ_year = df_pivot["Occupied Housing Units"].idxmax()
max_occ_value = df_pivot["Occupied Housing Units"].max()
print("\n                          Year   Value")
print(f"{'Vacant Housing Units':<25} {max_year}    {max_value}")
print(f"{'Occupied Housing Units':<25} {max_occ_year}    {max_occ_value}")

harmonic_mean_total = hmean(df_pivot["Total Housing Units"])
print("\nHarmonic Mean of Total Housing Units:", round(harmonic_mean_total, 2))
```

Develop a advance Python program to manage inventory records for a prodcut managing company. The system needs functions for adding, deleting and changing products and basic search options. Your system should use CSV files to automate restocking and produce reports that show item values alongside category breakdowns. Manage edge cases and corrupted CSV files. Finally maintaine 50 products of different categories in inventory.

```python
import csv
import os
import random
import logging
from collections import defaultdict

# Configure logging
logging.basicConfig(filename="inventory.log", level=logging.INFO, format="%(asctime)s - %(levelname)s - %(message)s")

# CSV File Path
INVENTORY_FILE = "inventory.csv"

# Default categories
CATEGORIES = ["Electronics", "Clothing", "Groceries", "Furniture", "Books"]

# Sample Inventory Data
INITIAL_PRODUCTS = [
    {"id": i, "name": f"Product-{i}", "category": random.choice(CATEGORIES), "quantity": random.randint(5, 50), "price": round(random.uniform(5, 500), 2)}
    for i in range(1, 51)
]

def initialize_inventory():
    """Initialize inventory CSV with sample data if not exists."""
    if not os.path.exists(INVENTORY_FILE):
        with open(INVENTORY_FILE, mode='w', newline='') as file:
            writer = csv.DictWriter(file, fieldnames=["id", "name", "category", "quantity", "price"])
            writer.writeheader()
            writer.writerows(INITIAL_PRODUCTS)
        logging.info("Initialized inventory with 50 products.")

def read_inventory():
    """Read inventory data from CSV file and handle corruption."""
    try:
        with open(INVENTORY_FILE, mode='r') as file:
            reader = csv.DictReader(file)
            return [row for row in reader]
    except Exception as e:
        logging.error(f"Error reading inventory: {e}")
        return []

def write_inventory(data):
    """Write inventory data to CSV file."""
    with open(INVENTORY_FILE, mode='w', newline='') as file:
        writer = csv.DictWriter(file, fieldnames=["id", "name", "category", "quantity", "price"])
        writer.writeheader()
        writer.writerows(data)

def add_product(name, category, quantity, price):
    """Add a new product to inventory."""
    if category not in CATEGORIES:
        print("Invalid category!")
        return
    inventory = read_inventory()
    new_id = max([int(item['id']) for item in inventory], default=0) + 1
    new_product = {"id": new_id, "name": name, "category": category, "quantity": quantity, "price": price}
    inventory.append(new_product)
    write_inventory(inventory)
    logging.info(f"Added product: {new_product}")

def delete_product(product_id):
    """Delete a product from inventory by ID."""
    inventory = read_inventory()
    updated_inventory = [item for item in inventory if int(item['id']) != product_id]
    if len(updated_inventory) == len(inventory):
        print("Product not found!")
        return
    write_inventory(updated_inventory)
    logging.info(f"Deleted product with ID: {product_id}")

def update_product(product_id, name=None, category=None, quantity=None, price=None):
    """Update product details in inventory."""
    inventory = read_inventory()
    for item in inventory:
        if int(item['id']) == product_id:
            if name: item['name'] = name
            if category and category in CATEGORIES: item['category'] = category
            if quantity is not None: item['quantity'] = quantity
            if price is not None: item['price'] = price
            write_inventory(inventory)
            logging.info(f"Updated product: {item}")
            return
    print("Product not found!")

def search_product(keyword):
    """Search products by keyword."""
    inventory = read_inventory()
    results = [item for item in inventory if keyword.lower() in item['name'].lower()]
    return results

def generate_report():
    """Generate a report of inventory value and category breakdown."""
    inventory = read_inventory()
    total_value = sum(float(item['quantity']) * float(item['price']) for item in inventory)
    category_count = defaultdict(int)
    for item in inventory:
        category_count[item['category']] += int(item['quantity'])

    print(f"Total Inventory Value: ${total_value:.2f}")
    print("Category Breakdown:")
    for category, count in category_count.items():
        print(f"  {category}: {count} items")
    logging.info("Generated inventory report.")

def restock_products(threshold=10, restock_amount=20):
    """Automatically restock products with low quantity."""
    inventory = read_inventory()
    for item in inventory:
        if int(item['quantity']) < threshold:
            item['quantity'] = str(int(item['quantity']) + restock_amount)
            logging.info(f"Restocked product {item['id']} ({item['name']}) to {item['quantity']} units.")
    write_inventory(inventory)

def main():
    initialize_inventory()
    while True:
        print("\nInventory Management System")
        print("1. Add Product")
        print("2. Delete Product")
        print("3. Update Product")
        print("4. Search Product")
        print("5. Generate Report")
        print("6. Restock Products")
        print("7. Exit")
        choice = input("Enter choice: ")

        if choice == "1":
            name = input("Enter product name: ")
            category = input(f"Enter category ({', '.join(CATEGORIES)}): ")
            quantity = int(input("Enter quantity: "))
            price = float(input("Enter price: "))
            add_product(name, category, quantity, price)
        elif choice == "2":
            product_id = int(input("Enter product ID to delete: "))
            delete_product(product_id)
        elif choice == "3":
            product_id = int(input("Enter product ID to update: "))
            name = input("Enter new name (or leave blank): ") or None
            category = input("Enter new category (or leave blank): ") or None
            quantity = input("Enter new quantity (or leave blank): ")
            price = input("Enter new price (or leave blank): ")
            update_product(product_id, name, category, int(quantity) if quantity else None, float(price) if price else None)
        elif choice == "4":
            keyword = input("Enter product name keyword: ")
            results = search_product(keyword)
            print(results if results else "No products found.")
        elif choice == "5":
            generate_report()
        elif choice == "6":
            restock_products()
        elif choice == "7":
            print("Exiting system.")
            break
        else:
            print("Invalid choice! Try again.")

if __name__ == "__main__":
    main()
```

```python
import csv
import os
import random
import logging
import re
from collections import defaultdict

# Configure logging
logging.basicConfig(filename="inventory.log", level=logging.INFO, format="%(asctime)s - %(levelname)s - %(message)s")

# CSV File Path
INVENTORY_FILE = "inventory.csv"

# Default categories
CATEGORIES = ["Electronics", "Clothing", "Groceries", "Furniture", "Books"]

# Sample Inventory Data
INITIAL_PRODUCTS = [
    {"id": i, "name": f"Product-{i}", "category": random.choice(CATEGORIES), "quantity": random.randint(5, 50), "price": round(random.uniform(5, 500), 2)}
    for i in range(1, 51)
]

# Maximum allowable quantity to prevent overflow
MAX_QUANTITY = 100000

# In-memory inventory data for faster operations
inventory_cache = {}

def initialize_inventory():
    """Initialize inventory CSV with sample data if not exists."""
    if not os.path.exists(INVENTORY_FILE):
        with open(INVENTORY_FILE, mode='w', newline='') as file:
            writer = csv.DictWriter(file, fieldnames=["id", "name", "category", "quantity", "price"])
            writer.writeheader()
            writer.writerows(INITIAL_PRODUCTS)
        logging.info("Initialized inventory with 50 products.")

    # Load data into memory for faster access
    load_inventory()

def load_inventory():
    """Load inventory data into memory for faster access."""
    global inventory_cache
    try:
        with open(INVENTORY_FILE, mode='r') as file:
            reader = csv.DictReader(file)
            inventory_cache = {int(row['id']): row for row in reader}
    except Exception as e:
        logging.error(f"Error reading inventory: {e}")
        print(f"Warning: Unable to load inventory. Some functionality may be limited. Error: {e}")

def write_inventory():
    """Write inventory data to CSV file."""
    try:
        with open(INVENTORY_FILE, mode='w', newline='') as file:
            writer = csv.DictWriter(file, fieldnames=["id", "name", "category", "quantity", "price"])
            writer.writeheader()
            writer.writerows(inventory_cache.values())
    except Exception as e:
        logging.error(f"Error writing inventory: {e}")
        print(f"Warning: Unable to save inventory changes. Error: {e}")

def is_valid_product_name(name):
    """Validate product name to ensure it only contains alphanumeric characters, spaces, or hyphens."""
    return bool(re.match("^[A-Za-z0-9\s\-]+$", name))

def get_non_empty_input(prompt):
    """Get user input that is not empty."""
    while True:
        user_input = input(prompt).strip()
        if user_input:
            return user_input
        else:
            print("Input cannot be empty. Please try again.")

def get_positive_integer_input(prompt):
    """Get a positive integer input."""
    while True:
        try:
            user_input = int(input(prompt).strip())
            if user_input > 0:
                return user_input
            else:
                print("Please enter a positive number.")
        except ValueError:
            print("Invalid input! Please enter a valid number.")

def get_positive_float_input(prompt):
    """Get a positive float input."""
    while True:
        try:
            user_input = float(input(prompt).strip())
            if user_input > 0:
                return user_input
            else:
                print("Please enter a positive number.")
        except ValueError:
            print("Invalid input! Please enter a valid number.")

def add_product(name, category, quantity, price):
    """Add a new product to inventory."""
    name = name.strip()  # Remove extra spaces
    if not is_valid_product_name(name):
        print("Invalid product name! Only letters, numbers, spaces, and hyphens are allowed.")
        return
    if any(item['name'].lower() == name.lower() for item in inventory_cache.values()):
        print("Product with this name already exists!")
        return
    if category not in CATEGORIES:
        print("Invalid category!")
        return
    if quantity <= 0 or quantity > MAX_QUANTITY:
        print(f"Quantity must be greater than zero and less than or equal to {MAX_QUANTITY}!")
        return
    if price <= 0:
        print("Price must be greater than zero!")
        return
    new_id = max(inventory_cache.keys(), default=0) + 1
    new_product = {"id": new_id, "name": name, "category": category, "quantity": quantity, "price": price}
    inventory_cache[new_id] = new_product
    write_inventory()
    logging.info(f"Added product: {new_product}")
    print("Product added successfully!")

def delete_product(product_id):
    """Delete a product from inventory by ID."""
    if product_id in inventory_cache:
        del inventory_cache[product_id]
        write_inventory()
        logging.info(f"Deleted product with ID: {product_id}")
        print("Product deleted successfully!")
    else:
        print("Product not found! No changes made.")

def update_product(product_id, name=None, category=None, quantity=None, price=None):
    """Update product details in inventory."""
    if product_id not in inventory_cache:
        print("Product not found! No changes made.")
        return

    product = inventory_cache[product_id]
    if name:
        name = name.strip()  # Remove extra spaces
        if not is_valid_product_name(name):
            print("Invalid product name! Only letters, numbers, spaces, and hyphens are allowed.")
            return
        if any(prod['name'].lower() == name.lower() and int(prod['id']) != product_id for prod in inventory_cache.values()):
            print("Product with this name already exists!")
            return
        product['name'] = name
    if category and category in CATEGORIES:
        product['category'] = category
    if quantity is not None:
        if quantity <= 0 or quantity > MAX_QUANTITY:
            print(f"Quantity must be greater than zero and less than or equal to {MAX_QUANTITY}!")
            return
        product['quantity'] = quantity
    if price is not None:
        if price <= 0:
            print("Price must be greater than zero!")
            return
        product['price'] = price

    write_inventory()
    logging.info(f"Updated product: {product}")
    print("Product updated successfully!")

def search_product(keyword):
    """Search products by keyword."""
    keyword = keyword.strip()  # Remove extra spaces
    results = [item for item in inventory_cache.values() if keyword.lower() in item['name'].lower()]

    if results:
        for product in results:
            print(f"ID: {product['id']}, Name: {product['name']}, Category: {product['category']}, Quantity: {product['quantity']}, Price: ${product['price']}")
    else:
        print("No products found!")

def generate_report():
    """Generate a report of inventory value and category breakdown."""
    total_value = sum(float(item['quantity']) * float(item['price']) for item in inventory_cache.values())
    category_count = defaultdict(int)
    for item in inventory_cache.values():
        category_count[item['category']] += int(item['quantity'])

    print(f"Total Inventory Value: ${total_value:.2f}")
    print("Category Breakdown:")
    for category, count in category_count.items():
        print(f"  {category}: {count} items")
    logging.info("Generated inventory report.")

def restock_products(threshold=10, restock_amount=20):
    """Automatically restock products with low quantity."""
    if threshold < 0:
        print("Threshold cannot be negative! Using default value of 10.")
        threshold = 10

    restocked = False
    for item in inventory_cache.values():
        if int(item['quantity']) < threshold:
            item['quantity'] = str(int(item['quantity']) + restock_amount)
            logging.info(f"Restocked product {item['id']} ({item['name']}) to {item['quantity']} units.")
            restocked = True

    if restocked:
        write_inventory()
        print("Low-stock products have been restocked.")
    else:
        print("No products needed restocking.")

def main():
    initialize_inventory()
    while True:
        print("\nInventory Management System")
        print("1. Add Product")
        print("2. Delete Product")
        print("3. Update Product")
        print("4. Search Product")
        print("5. Generate Report")
        print("6. Restock Products")
        print("7. Exit")
        choice = input("Enter choice: ").strip()

        if choice == "1":
            name = get_non_empty_input("Enter product name: ")
            category = get_non_empty_input(f"Enter category ({', '.join(CATEGORIES)}): ")
            quantity = get_positive_integer_input("Enter quantity: ")
            price = get_positive_float_input("Enter price: ")
            add_product(name, category, quantity, price)

        elif choice == "2":
            product_id = get_positive_integer_input("Enter product ID to delete: ")
            delete_product(product_id)

        elif choice == "3":
            product_id = get_positive_integer_input("Enter product ID to update: ")
            if product_id not in inventory_cache:
                print("Invalid Product ID! Please enter a valid product ID.")
                continue

            name = get_non_empty_input("Enter new name (or leave blank): ") or None
            category = get_non_empty_input("Enter new category (or leave blank): ") or None
            quantity = input("Enter new quantity (or leave blank): ")
            price = input("Enter new price (or leave blank): ")

            update_product(product_id, name, category, int(quantity) if quantity else None, float(price) if price else None)

        elif choice == "4":
            keyword = get_non_empty_input("Enter product name keyword: ")
            search_product(keyword)

        elif choice == "5":
            generate_report()

        elif choice == "6":
            try:
                threshold = int(input("Enter low-stock threshold (default 10): ") or 10)
                restock_amount = int(input("Enter restock amount (default 20): ") or 20)
                restock_products(threshold, restock_amount)
            except ValueError:
                print("Invalid input! Please enter numeric values.")

        elif choice == "7":
            print("Exiting system.")
            break

        else:
            print("Invalid choice! Try again.")

if __name__ == "__main__":
    main()
```

```python
import random
from datetime import datetime, timedelta

class Flight:
    def __init__(self, flight_number, origin, destination, departure_time, arrival_time, capacity, base_price):
        self.flight_number = flight_number
        self.origin = origin
        self.destination = destination
        self.departure_time = departure_time
        self.arrival_time = arrival_time
        self.capacity = capacity
        self.available_seats = capacity
        self.base_price = base_price
        self.reservations = {}
        self.waitlist = []

    def reserve_seat(self, passenger_name):
        if self.available_seats > 0:
            seat_number = self.capacity - self.available_seats + 1
            self.reservations[passenger_name] = seat_number
            self.available_seats -= 1
            print(f"Seat {seat_number} reserved for {passenger_name} on flight {self.flight_number}.")
        else:
            if passenger_name not in self.waitlist:
                self.waitlist.append(passenger_name)
                print(f"Flight full. {passenger_name} added to waitlist.")
            else:
                print(f"{passenger_name} is already on the waitlist.")

    def release_seat(self, passenger_name):
        if passenger_name in self.reservations:
            del self.reservations[passenger_name]
            self.available_seats += 1
            print(f"Seat released for {passenger_name} on flight {self.flight_number}.")
            if self.waitlist:
                next_passenger = self.waitlist.pop(0)
                self.reserve_seat(next_passenger)
        else:
            print(f"No reservation found for {passenger_name}.")

    def get_price(self):
        dynamic_price = self.base_price + (self.base_price * 0.1 * (self.capacity - self.available_seats) / self.capacity)
        return round(dynamic_price, 2)

    def __str__(self):
        return f"{self.flight_number}: {self.origin} -> {self.destination} ({self.available_seats}/{self.capacity} seats available)"

class FlightBookingSystem:
    def __init__(self):
        self.flights = {}

    def add_flight(self, flight):
        if flight.flight_number not in self.flights:
            self.flights[flight.flight_number] = flight
        else:
            print(f"Flight {flight.flight_number} already exists. Duplicate flights are not allowed.")

    def find_flights(self, origin, destination):
        return [flight for flight in self.flights.values() if flight.origin == origin and flight.destination == destination]

    def reserve_seat(self, flight_number, passenger_name):
        if flight_number in self.flights:
            self.flights[flight_number].reserve_seat(passenger_name)
        else:
            print(f"Flight {flight_number} does not exist. Cannot proceed with the reservation.")

    def generate_report(self):
        print("\n--- Flight Performance Report ---")
        for flight in self.flights.values():
            print(f"{flight.flight_number}: {len(flight.reservations)} seats booked, {flight.available_seats} available, Current Price: ${flight.get_price()}")

# Sample flight data
flight_system = FlightBookingSystem()

# Generate 50 unique flights
cities = ["New York", "Los Angeles", "Chicago", "Houston", "Miami", "San Francisco", "Seattle", "Dallas", "Boston", "Denver"]
for i in range(50):
    origin, destination = random.sample(cities, 2)
    departure = datetime.now() + timedelta(days=random.randint(1, 30))
    arrival = departure + timedelta(hours=random.randint(2, 6))
    capacity = random.randint(100, 300)
    base_price = random.randint(100, 500)
    flight_number = f"FL{i+100}"
    flight = Flight(flight_number, origin, destination, departure, arrival, capacity, base_price)
    flight_system.add_flight(flight)

# Testing system functionality
while True:
    try:
        flight_system.generate_report()
        found_flights = flight_system.find_flights("New York", "Los Angeles")
        if found_flights:
            flight = found_flights[0]
            flight_system.reserve_seat(flight.flight_number, "Alice")
            flight_system.reserve_seat(flight.flight_number, "Bob")
            flight.release_seat("Alice")
            flight_system.generate_report()
        else:
            print("No flights found.")
        break
    except Exception as e:
        print(f"An error occurred: {e}. Please try again.")
```
